```python
import snscrape.modules.twitter as sntwitter
import  pandas  as pd
tweets=[]
limit=2500
```

```python
query="EV (buy OR exchange) min_faves:20 min_retweets:20 lang:en   "

for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    if len(tweets)==limit:
        break 
    else:
        tweets.append([tweet.date,tweet.user.username,tweet.content,tweet.retweetCount,tweet.likeCount])
df = pd.DataFrame(tweets,columns=['date','username','content','retweet_count','like_count'])
    #print(vars(tweet)
    #break 
    
```
